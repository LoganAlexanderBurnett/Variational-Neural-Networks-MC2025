```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def load_std_to_arrays(filename: str):
    # Load the CSV into a DataFrame
    df = pd.read_csv(filename)

    # Convert columns back into three separate arrays using column numbers
    array1 = df.iloc[:, 0:3].to_numpy()
    array2 = df.iloc[:, 3:6].to_numpy()

    return array1, array2

def load_var_to_arrays(filename: str):
    # Load the CSV into a DataFrame
    df = pd.read_csv(filename)

    # Convert columns back into three separate arrays using column numbers
    array1 = df.iloc[:, 0:3].to_numpy()
    array2 = df.iloc[:, 3:6].to_numpy()
    array3 = df.iloc[:, 6:9].to_numpy()

    return array1, array2, array3

# Return arrays for std and var
stdGRUtest_actual, stdGRUtest_preds = load_std_to_arrays('./StandardGRUTest.csv')
vGRUtest_actual, vGRUtest_mean, vGRUtest_sigma = load_var_to_arrays('./vGRUTest.csv')

# Verify the shape
print(stdGRUtest_actual.shape)  # (6464, 3)
print(stdGRUtest_preds.shape)  # (6464, 3)
print(vGRUtest_actual.shape)  # (6464, 3)
print(vGRUtest_mean.shape)  # (6464, 3)
print(vGRUtest_sigma.shape)  # (6464, 3)
```

```python
labels = ['Solar Power', 'Load Power', 'Wind Power']

# Plot predictions vs actuals
for i, label in zip(range(0,3), labels):
    plt.figure(figsize=(16,6))
    plt.plot(stdGRUtest_actual[:, i], label='Actual', linestyle='-', color='blue')
    plt.plot(stdGRUtest_preds[:, i], label='Prediction', color='r', linestyle='--')
    
    plt.ylabel(label)
    plt.xlabel('Time (minutes)')
    plt.legend()
    plt.grid(True)
    plt.show()


```

```python
labels = ['Solar Power', 'Load Power', 'Wind Power']

def plot_variational(true_values, mean_predictions, ci, output_index, mean_color, ci_color): 
    # Plot predictions vs actuals
    
    # Plot true values (ground truth)
    # plt.plot(true_values[:, output_index], label='Actual', color='blue', linewidth=2)
    
    # Plot predicted mean
    plt.plot(mean_predictions[:, output_index], label='vGRU Mean', color=mean_color, linestyle='--', linewidth=1)
    
    # Plot confidence intervals (mean +/- confidence interval)
    plt.fill_between(
        np.arange(mean_predictions.shape[0]),
        mean_predictions[:, output_index] - ci[:, output_index],
        mean_predictions[:, output_index] + ci[:, output_index],
        color=ci_color, alpha=0.2, label='vGRU 95% Confidence Interval'
    )

    plt.ylabel(labels[output_index])
    plt.xlabel('Time (minutes)')
    plt.legend()
    plt.grid(True)
    # plt.show()
    
```

```python
plot_variational(vGRUtest_actual, vGRUtest_mean, vGRUtest_sigma, 0, 'orange', 'green')
```

```python
labels = ['Solar Power', 'Load Power', 'Wind Power']

# Plot predictions vs actuals
for i, label in zip(range(0,1), labels):
    plt.figure(figsize=(10,3))
    
    plt.plot(stdGRUtest_actual[:, i], label='Actual', linestyle='-', color='k')
    plt.plot(stdGRUtest_preds[:, i], label='GRU', color='b', linestyle='--')

    plot_variational(vGRUtest_actual, vGRUtest_mean, vGRUtest_sigma, 0, 'r', 'r')
    
    plt.ylabel(label)
    plt.xlabel('Time (minutes)')
    plt.legend(loc='upper right', framealpha=0.1, ncol=4)
    plt.grid(True)
    plt.xlim(0, 1300)
    plt.savefig("GRU_test_preds_solarREVISION.png", dpi=300, bbox_inches='tight')
    plt.show()

```

```python
# Plot predictions vs actuals
for i, label in zip(range(1,2), labels):
    plt.figure(figsize=(16,6))
    
    plt.plot(stdGRUtest_actual[:, i], label='Actual', linestyle='-', color='k')
    plt.plot(stdGRUtest_preds[:, i], label='LSTM', color='b', linestyle='--')

    plot_variational(vGRUtest_actual, vGRUtest_mean, vGRUtest_sigma, 1, 'r', 'r')
    
    plt.ylabel('Load Power')
    plt.xlabel('Time (minutes)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Plot predictions vs actuals
for i, label in zip(range(2,3), labels):
    plt.figure(figsize=(10,3))
    
    plt.plot(stdGRUtest_actual[:, i], label='Actual', linestyle='-', color='k')
    plt.plot(stdGRUtest_preds[:, i], label='GRU', color='b', linestyle='--')

    plot_variational(vGRUtest_actual, vGRUtest_mean, vGRUtest_sigma, 2, 'r', 'r')
    
    plt.ylabel('Wind Power')
    plt.xlabel('Time (minutes)')
    plt.legend()
    plt.grid(True)
    plt.xlim(0, 1300)
    plt.legend(loc='upper right', framealpha=0.1, ncol=4)
    plt.savefig("GRU_test_preds_windREVISION.png", dpi=300, bbox_inches='tight')
    plt.show()
```
